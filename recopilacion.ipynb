## importación  de librerias de python

Este código está diseñado para montar Google Drive en un entorno de Google Colab, configurar la API de YouTube, extraer transcripciones de videos en una lista de reproducción de YouTube y guardarlas en archivos de texto en Google Drive. Además, registra detalles sobre cada transcripción en un archivo CSV llamado registro_transcripciones.csv el cual guarda nombre del archivo, su
longitud, fecha y sitio de donde se descargó

```python
import csv
import os
import re
from datetime import datetime
from googleapiclient.discovery import build
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled, NoTranscriptFound

# Ruta donde quieres guardar los archivos en tu equipo local
ruta_guardado_local = './Mananeras'
contador = 0  # Inicializamos el contador

# Crea la carpeta si no existe
if not os.path.exists(ruta_guardado_local):
    os.makedirs(ruta_guardado_local)

# Configuración de la API de YouTube
API_KEY = 'AIzaSyDEM4jEG6x1zJIvcX8u-Ffjkb3zXBtxGNQ'
youtube = build('youtube', 'v3', developerKey=API_KEY)

# Función para extraer el playlist_id desde la URL del primer video
def extraer_playlist_id(url_video):
    match = re.search(r"[&?]list=([a-zA-Z0-9_-]+)", url_video)
    if match:
        return match.group(1)
    else:
        print("No se encontró un playlist_id en la URL proporcionada.")
        return None

# Función para obtener la lista completa de videos de la lista de reproducción
def obtener_videos_playlist(playlist_id):
    videos = []
    request = youtube.playlistItems().list(
        part="snippet",
        playlistId=playlist_id,
        maxResults=50
    )
    while request:
        response = request.execute()
        for item in response['items']:
            video_id = item['snippet']['resourceId']['videoId']
            video_title = item['snippet']['title']
            fecha_publicacion = item['snippet']['publishedAt']
            videos.append((video_id, video_title, fecha_publicacion))
        request = youtube.playlistItems().list_next(request, response)
    return videos

# Función para limpiar el nombre del archivo eliminando caracteres no permitidos
def limpiar_nombre_archivo(nombre):
    # Reemplaza caracteres no permitidos con un guion bajo
    return re.sub(r'[<>:"/\\|?*]', '_', nombre)

# Función para procesar cada video de la lista y guardar la transcripción
def procesar_lista_reproduccion(url_video):
    global contador  # Declaramos que usaremos la variable global
    playlist_id = extraer_playlist_id(url_video)
    if not playlist_id:
        return

    videos = obtener_videos_playlist(playlist_id)
    ruta_csv = os.path.join(ruta_guardado_local, 'registro_transcripciones.csv')

    with open(ruta_csv, 'a', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)

        # Escribir encabezado si el archivo está vacío
        if os.stat(ruta_csv).st_size == 0:
            writer.writerow([
                "Nombre Archivo", "Longitud (palabras)", "Fecha Publicación",
                "Fecha Consulta", "Hora Consulta", "Sitio de Origen"
            ])

        for video_id, video_title, fecha_publicacion in videos:
            try:
                transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=['es'])
                lineas = [item['text'] for item in transcript]

                # Combinar todas las líneas de la transcripción en un solo texto
                contenido_transcripcion = " ".join(lineas)

                # Guardar transcripción en archivo de texto en la carpeta local
                nombre_archivo = limpiar_nombre_archivo(f"{video_title}.txt")
                ruta_archivo_txt = os.path.join(ruta_guardado_local, nombre_archivo)
                with open(ruta_archivo_txt, 'w', encoding='utf-8') as f:
                    f.write(contenido_transcripcion)

                # Registrar toda la transcripción como una sola entrada en el CSV
                fecha_consulta = datetime.now().strftime("%Y-%m-%d")
                hora_consulta = datetime.now().strftime("%H:%M:%S")
                writer.writerow([
                    nombre_archivo, len(contenido_transcripcion.split()), fecha_publicacion,
                    fecha_consulta, hora_consulta, f"https://www.youtube.com/watch?v={video_id}"
                ])

                # Incrementar el contador
                contador += 1
                print(f"Transcripción de '{video_title}' guardada en {nombre_archivo}")
                
            except TranscriptsDisabled:
                print(f"Transcripción no disponible para el video: {video_title}")
            except NoTranscriptFound:
                print(f"No se encontró transcripción para el video: {video_title}")
            except Exception as e:
                print(f"Error al procesar el video {video_title}: {e}")

# Ejemplo de uso
url_video = 'https://www.youtube.com/watch?v=V8yYRcmpLkc&list=PL-wEE8VmWaJ3BoPk-jxOrjOp711iP_Oqg'
procesar_lista_reproduccion(url_video)
print(f"NÚMERO DE VIDEOS RECUPERADOS: {contador}")
```

## Ordena los archivos dentro de la tabla csv por fecha

```python
import csv
from datetime import datetime

# Ruta al archivo CSV
ruta_csv = 'Mananeras/registro_transcripciones.csv'

# Leer y ordenar el archivo CSV por la columna "Fecha Publicación"
def ordenar_csv_por_fecha(ruta_csv):
    with open(ruta_csv, 'r', encoding='utf-8') as archivo_csv:
        reader = csv.reader(archivo_csv)
        encabezado = next(reader)  # Guardar la primera fila como encabezado

        # Convertir las filas en una lista y ordenar por la fecha de publicación (índice 2 en este caso)
        filas_ordenadas = sorted(reader, key=lambda row: datetime.fromisoformat(row[2]))

    # Guardar el archivo ordenado
    with open(ruta_csv, 'w', newline='', encoding='utf-8') as archivo_csv:
        writer = csv.writer(archivo_csv)
        writer.writerow(encabezado)  # Escribir el encabezado nuevamente
        writer.writerows(filas_ordenadas)  # Escribir las filas ordenadas

    print("Archivo CSV ordenado por fecha de publicación.")

# Ejecutar la función
ordenar_csv_por_fecha(ruta_csv)
```

```python
import os
import re
from nltk.corpus import stopwords

# Configurar rutas
ruta_guardado_local = './Mananeras'
ruta_guardado_limpio = './Mananeras_limpio'

# Crear la carpeta para archivos limpios si no existe
if not os.path.exists(ruta_guardado_limpio):
    os.makedirs(ruta_guardado_limpio)

# Obtener lista de palabras vacías en español
# Asegúrate de descargar las stopwords de NLTK previamente:
# import nltk
# nltk.download('stopwords')
stopwords_es = set(stopwords.words('spanish'))

# Función para limpiar texto
def limpiar_texto(texto):
    # Eliminar caracteres que no sean letras, números o espacios
    texto = re.sub(r'[^\wáéíóúñüÁÉÍÓÚÑÜ\s]', '', texto)
    # Convertir a minúsculas
    texto = texto.lower()
    # Eliminar palabras vacías
    palabras = texto.split()
    palabras_filtradas = [palabra for palabra in palabras if palabra not in stopwords_es]
    # Unir el texto limpio
    texto_limpio = ' '.join(palabras_filtradas)
    return texto_limpio

# Función para procesar todos los archivos de texto en la carpeta
def limpiar_archivos_en_carpeta(ruta_carpeta_origen, ruta_carpeta_destino):
    archivos = [f for f in os.listdir(ruta_carpeta_origen) if f.endswith('.txt')]
    for archivo in archivos:
        ruta_archivo_origen = os.path.join(ruta_carpeta_origen, archivo)
        with open(ruta_archivo_origen, 'r', encoding='utf-8') as f:
            contenido = f.read()
        
        # Limpiar el contenido
        contenido_limpio = limpiar_texto(contenido)
        
        # Guardar el contenido limpio en la carpeta destino
        ruta_archivo_destino = os.path.join(ruta_carpeta_destino, archivo)
        with open(ruta_archivo_destino, 'w', encoding='utf-8') as f:
            f.write(contenido_limpio)
        
        print(f"Archivo limpio guardado como: {ruta_archivo_destino}")

# Ejecutar limpieza sobre los documentos recuperados
limpiar_archivos_en_carpeta(ruta_guardado_local, ruta_guardado_limpio)
```

## Reconocimiento de palabras 


```python
import os
from spellchecker import SpellChecker

# Configurar rutas
ruta_guardado_limpio = './Mananeras_limpio'
ruta_guardado_correcciones = './Mananeras_correcciones'
ruta_palabras_no_reconocidas = './Palabras_no_reconocidas'

# Crear carpetas si no existen
os.makedirs(ruta_guardado_correcciones, exist_ok=True)
os.makedirs(ruta_palabras_no_reconocidas, exist_ok=True)

# Inicializar el corrector ortográfico para español
spell = SpellChecker(language='es')

# Función para corregir ortografía en un archivo
def corregir_archivo(ruta_origen, ruta_destino, ruta_palabras_no_reconocidas):
    with open(ruta_origen, 'r', encoding='utf-8') as f:
        contenido = f.read()

    palabras = contenido.split()
    palabras_corregidas = []
    palabras_no_reconocidas = []

    for palabra in palabras:
        if palabra in spell:
            palabras_corregidas.append(palabra)
        else:
            palabras_no_reconocidas.append(palabra)
            palabra_corregida = spell.correction(palabra)
            palabras_corregidas.append(palabra_corregida if palabra_corregida else palabra)

    # Guardar texto corregido
    with open(ruta_destino, 'w', encoding='utf-8') as f:
        f.write(' '.join(palabras_corregidas))

    # Guardar palabras no reconocidas
    if palabras_no_reconocidas:
        ruta_archivo_no_reconocidas = os.path.join(
            ruta_palabras_no_reconocidas,
            f"palabras_no_reconocidas_{os.path.basename(ruta_origen)}"
        )
        with open(ruta_archivo_no_reconocidas, 'w', encoding='utf-8') as f:
            f.write('\n'.join(set(palabras_no_reconocidas)))

    print(f"Archivo corregido guardado como: {ruta_destino}")
    print(f"Palabras no reconocidas guardadas en: {ruta_archivo_no_reconocidas}")

# Función para procesar archivos con reanudación
def procesar_archivos_correccion(ruta_carpeta_origen, ruta_carpeta_destino, ruta_palabras_no_reconocidas):
    archivos = [f for f in os.listdir(ruta_carpeta_origen) if f.endswith('.txt')]

    # Identificar archivos ya procesados
    archivos_procesados = set(os.listdir(ruta_carpeta_destino))

    for archivo in archivos:
        ruta_archivo_origen = os.path.join(ruta_carpeta_origen, archivo)
        ruta_archivo_destino = os.path.join(ruta_carpeta_destino, archivo)

        # Saltar si el archivo ya fue procesado
        if archivo in archivos_procesados:
            print(f"Archivo {archivo} ya procesado. Saltando...")
            continue

        try:
            corregir_archivo(ruta_archivo_origen, ruta_archivo_destino, ruta_palabras_no_reconocidas)
        except Exception as e:
            print(f"Error procesando {archivo}: {e}")
            break  # Detener el proceso en caso de error

# Ejecutar corrección ortográfica
procesar_archivos_correccion(ruta_guardado_limpio, ruta_guardado_correcciones, ruta_palabras_no_reconocidas)



```
